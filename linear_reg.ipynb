```python
import numpy as np
```

```python
x = np.random.uniform(0,10,10)
y = 7*x + 5
```

```python
x,y
```

```python
n = len(x)

Sxy = sum(x*y)- (1.0/n)*sum(x)*sum(y)

Sxx = sum(x**2) - (1.0/n)*sum(x)**2

beta1 = Sxy/Sxx

alpha1 = sum(y)/n - beta1*(sum(x)/n)

print(alpha1,beta1)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x.reshape(-1, 1),y)

alpha2 = lr.intercept_
beta2 = lr.coef_

print(alpha2, beta2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.scatter(x,y,c ='g')

x_ = np.linspace(min(x),max(x),100)
y_ = alpha1+ beta1*x_
plt.plot(x_,y_,'m--')

plt.legend(['Datas', 'LR line'])
plt.title("Linear Regression 1")
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y,c ='g')

x_ = np.linspace(min(x),max(x),100)
y_ = alpha2+ beta2*x_
plt.plot(x_,y_,'r--')

plt.legend(['Datas', 'LR line'])
plt.title("Linear Regression 1")
```
